## 6. Create a spending flag

6. The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all their orders using the following criteria:
   - If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
   - If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”

```python
#Create column with average price orders
ords_prods_merge['average_price_orders'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge[['user_id', 'average_price_orders']].head(100)
```

```python
# Create spending_flag
ords_prods_merge.loc[ords_prods_merge['average_price_orders'] < 10, 'spending_flag'] = 'Low spender'
ords_prods_merge.loc[(ords_prods_merge['average_price_orders'] >= 10), 'spending_flag'] = 'High spender'
```

```python
# Check value counts on the spending_flag
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
ords_prods_merge.head(100)
```

As a note, there are no high spenders because the average is always lower than 10. This might have to do with the fact that I corrected some price values previously, since there were some outliers.

## 7. Create an order frequency flag

7. In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:
- If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
- If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
- If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”

```python
#Crerate column with median values of ordering behavior
ords_prods_merge['median_ordering_behavior'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge.head(100)
```

```python
#Create order_frequency_flag
ords_prods_merge.loc[ords_prods_merge['median_ordering_behavior'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
ords_prods_merge.loc[(ords_prods_merge['median_ordering_behavior'] > 10) & (ords_prods_merge['median_ordering_behavior'] <= 20), 'order_frequency_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['median_ordering_behavior'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
ords_prods_merge.head(100)
```

```python
# Check value counts on the spending_flag
ords_prods_merge['order_frequency_flag'].value_counts(dropna = False)
```

```python
#Check NaN values
missing_values = ords_prods_merge[ords_prods_merge['order_frequency_flag'].isnull() == True]
```

```python
missing_values
```

Makes sense since there are missing values in the days_since_prior_order that match clients who have never purchased before.
